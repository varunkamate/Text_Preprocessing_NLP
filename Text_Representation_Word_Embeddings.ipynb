# Bag of words

```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({"text":["people watch ironman",
                         "ironman watch ironman",
                         "people write comment",
                          "ironman write comment"],"output":[1,0,1,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
a=cv.fit_transform(df["text"])
```

vocabulary

```python
print(cv.vocabulary_)
```

```python
a.toarray()
```

```python
print(a[0].toarray())
print(a[1].toarray())
print(a[2].toarray())
```

```python
# new
cv.transform(['ironman watch ironman']).toarray()
```

#  N-grams

```python
df=pd.DataFrame({"text":["people watch ironman",
                         "ironman watch ironman",
                         "people write comment",
                          "ironman write comment"],"output":[1,0,1,0]})
```

```python
df
```

```python
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
b=cv.fit_transform(df["text"])
```

```python
print(cv.vocabulary_)
```

```python
print(b[0].toarray())
print(b[1].toarray())
print(b[2].toarray())
```

# TF-IDF (Term frequency- Inverse document frequency)

```python
df=pd.DataFrame({"text":["people watch ironman",
                         "ironman watch ironman",
                         "people write comment",
                          "ironman write comment"],"output":[1,0,1,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfid= TfidfVectorizer()
```

```python
arr = tfid.fit_transform(df['text']).toarray()
```

```python
arr
```

```python
print(tfid.idf_)
```
